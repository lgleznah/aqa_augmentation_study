```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.layers import RandomBrightness, RandomContrast, RandomFlip, RandomTranslation, RandomRotation, RandomZoom
```

```python
def parse_image(filename):
    image = tf.io.read_file(filename)
    image = tf.io.decode_jpeg(image)
    
    return image
```

```python
img = parse_image('great_photo.jpg')
```

```python
bad_img_1 = RandomBrightness(1.0)(img)
```

```python
bad_img_2 = RandomContrast(1.0)(img)
```

```python
bad_img_3 = RandomFlip()(img)
```

```python
bad_img_4 = RandomTranslation(1.0, 1.0, fill_mode='constant', fill_value=255.0)(img)
```

```python
bad_img_5 = RandomRotation(1.0, fill_mode='constant', fill_value=255.0)(img)
```

```python
bad_img_6 = RandomZoom(1.0, 1.0, fill_mode='constant', fill_value=255.0)(img)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(24,10))

for ax in axes.flatten():
    ax.set_xticks([])
    ax.set_yticks([])
    ax.axis('off')

axes[0,1].imshow(img)
axes[1,0].imshow(bad_img_1 / 255)
axes[1,1].imshow(bad_img_5 / 255)
axes[1,2].imshow(bad_img_6 / 255)

plt.tight_layout()
plt.subplots_adjust(hspace=0.05)
plt.savefig('image_destruction.pdf')
```
